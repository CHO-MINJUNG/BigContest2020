```python
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import array

main = 'C:/Users/chomjung/Downloads/2020빅콘테스트'
EDA_address = '/EDA_세이버 매트릭스_'
team_pitcher = pd.read_csv(main+"/팀투수"+EDA_address+'팀투수.csv')
pitcher = pd.read_csv(main+"/개인투수"+EDA_address+'개인투수.csv')
team_hitter = pd.read_csv(main+"/팀타자"+EDA_address+'팀타자.csv')
hitter = pd.read_csv(main+"/개인타자"+EDA_address+'개인타자.csv')
```

```python
def RC(row):
    previous = team_hitter[(team_hitter['시즌'] == row['시즌']) & (team_hitter['팀코드'] == row['팀코드']) & (team_hitter['일자'] < row['일자'])]
    if len(previous) == 0:
        row['RC'] = 0
        return row
    previous_sum = previous.sum()
    upA = previous_sum['안타']+previous_sum['4구']-previous_sum['도루실패']+previous_sum['사구']-previous_sum['병살타']
    upB = (previous_sum['단타']*1.125)+(1.69*previous_sum['2루타'])+(3.02*previous_sum['3루타'])+(3.73*previous_sum['홈런'])+ 0.29*(previous_sum['4구']-previous_sum['고4']+previous_sum['사구'])
    down = previous_sum['타수']+previous_sum['4구']+previous_sum['사구']+previous_sum['희비']+previous_sum['희타']
    if down == 0:
        row['RC'] = 0
        return row
    row['RC'] = ((2.4*down+upA)*(3*down+upB)/(9*down)) - (0.9*down)
    
    return row

def RC27(row):
    previous = team_hitter[(team_hitter['시즌'] == row['시즌']) & (team_hitter['팀코드'] == row['팀코드']) & (team_hitter['일자'] < row['일자'])]
    if len(previous) == 0:
        row['RC27'] = 0
        return row
    previous_sum = previous.sum()
    down = previous_sum['타수'] - previous_sum['안타']+previous_sum['병살타']+previous_sum['도루실패']+previous_sum['희타']+previous_sum['희비']
    if down == 0:
        row['RC27'] = 0
        return row
    row['RC27'] = (row['RC']/down)*27
    
    return row
```

```python
team_hitter = team_hitter.apply(RC, axis=1)
```

```python
team_hitter = team_hitter.apply(RC27, axis=1)
```
