```
teamhitterNC = teamhitter[teamhitter['팀코드']=='NC']
teamhitterNC.reset_index(drop=True)
scoreNC = teamhitterNC['득점']
del teamhitterNC['득점']
del teamhitterNC['팀코드']

teamhitterSK = teamhitter[teamhitter['팀코드']=='SK']
teamhitterSK.reset_index(drop=True)
scoreSK = teamhitterSK['득점']
del teamhitterSK['득점']

teamhitterKT = teamhitter[teamhitter['팀코드']=='KT']
teamhitterKT.reset_index(drop=True)
scoreKT = teamhitterKT['득점']
del teamhitterKT['득점']

teamhitterHT = teamhitter[teamhitter['팀코드']=='HT']
teamhitterHT.reset_index(drop=True)
scoreHT = teamhitterHT['득점']
del teamhitterHT['득점']

teamhitterLT = teamhitter[teamhitter['팀코드']=='LT']
teamhitterLT.reset_index(drop=True)
scoreLT = teamhitterLT['득점']
del teamhitterLT['득점']

teamhitterWO = teamhitter[teamhitter['팀코드']=='WO']
teamhitterWO.reset_index(drop=True)
scoreWO = teamhitterWO['득점']
del teamhitterWO['득점']

teamhitterOB = teamhitter[teamhitter['팀코드']=='OB']
teamhitterOB.reset_index(drop=True)
scoreOB = teamhitterOB['득점']
del teamhitterOB['득점']

teamhitterHH = teamhitter[teamhitter['팀코드']=='HH']
teamhitterHH.reset_index(drop=True)
scoreHH = teamhitterHH['득점']
del teamhitterHH['득점']

teamhitterLG = teamhitter[teamhitter['팀코드']=='LG']
teamhitterLG.reset_index(drop=True)
scoreLG = teamhitterLG['득점']
del teamhitterLG['득점']

teamhitterSS = teamhitter[teamhitter['팀코드']=='SS']
teamhitterSS.reset_index(drop=True)
scoreSS = teamhitterSS['득점']
del teamhitterSS['득점']
```

```
NCtrain =list( teamhitterNC.values.tolist() ) 
NCtrain_x = np.array(NCtrain)
NCtrain_y = np.array(scoreNC.values.tolist())
```

```
NCtrain_x = NCtrain_x.reshape(NCtrain_x.shape[0],NCtrain_x.shape[1],1)
NCtrain_x.shape
```

```
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, LSTM
model = Sequential()

model.add(LSTM(128, input_shape = (21,1), activation = 'relu'))
model.add(Dense(64))
model.add(Dense(1))

model.summary()

model.compile(loss='mse', optimizer = 'adam', metrics = ['mse'])
model.fit(NCtrain_x, NCtrain_y, epochs = 30, batch_size=9)
```

```
loss, acc = model.evaluate(NCtrain_x, NCtrain_y)
print("학습 후 모델의 loss: {:.5f}".format(loss))
print("학습 후 모델의 정확도: {:5.2f}%".format(100*acc))
```

```
def preprocess():
  # 1차_최종_공통 : 시즌, 홈어웨이, 경기간간격, 득점, 실점, 승패
  general = pd.read_csv('./gdrive/My Drive/빅콘테스트2020/1차_최종_데이터/1차_최종_공통.csv')
  general = general[general['팀코드']=='NC']
  general = general[['게임키', '경기간간격','득점','실점','결과']]
  general['시즌'] = general['게임키'].apply(lambda x:int(x[:4]))
  general['홈어웨이'] = general['게임키'].apply(lambda x:0 if x[-3:-1]=='NC' else 1)
  general=general.set_index('게임키')
  # 팀타자 : 기준팀 wRC
  hitter = teamhitter
  hitter_vs = hitter[hitter['상대팀코드']=='NC']
  hitter = hitter[hitter['팀코드']=='NC']
  hitter = hitter[['게임키', '상대최근RC']]
  hitter_vs = hitter_vs[['게임키', '상대최근RC']]
  hitter=hitter.set_index('게임키')
  hitter_vs=hitter_vs.set_index('게임키')
  hitter_vs=hitter_vs.add_prefix('vs_')
  df = pd.concat([general, hitter,hitter_vs], axis=1)
  df= df[['시즌','경기간간격','홈어웨이','상대최근RC','vs_상대최근RC','득점','실점','결과']]
  df['결과'] = df['결과'].apply(lambda x : 0 if x =='W' else 1)
  return df
```

```
def create_dataset(signal_data, look_back=1):
    dataX, dataY = [], []
    for i in range(len(signal_data)-look_back):
        dataX.append(signal_data[i:(i+look_back), 0])
        dataY.append(signal_data[i + look_back, 0])
    return np.array(dataX), np.array(dataY)
```
