```
hitter_seasons = [2016,2017,2018,2019,2020]
pitcher_seasons = [2016,2017,2018,2019,2020]
for i in range(5):
  hitter_seasons[i] = teamhitter[teamhitter["시즌"]==hitter_seasons[i]]
for i in range(5):
  hitter_seasons[i] =hitter_seasons[i].reset_index(drop=True)

for i in range(5):
  pitcher_seasons[i] = teampitcher[teampitcher["시즌"]==pitcher_seasons[i]]
for i in range(5):
  pitcher_seasons[i] =pitcher_seasons[i].reset_index(drop=True)
```

```
for i in range(5):
  hitter_seasons[i] = hitter_seasons[i].groupby('일자').sum()
for i in range(5):
  hitter_seasons[i] = hitter_seasons[i].reset_index()

for i in range(5):
  pitcher_seasons[i] = pitcher_seasons[i].groupby('일자').sum()
for i in range(5):      
  pitcher_seasons[i] = pitcher_seasons[i].reset_index()
```

```
for j in range(5):
  all_list = []
  for i in range(len(hitter_seasons[j])):
    if i == 0:
      all_list.append(hitter_seasons[j]['홈런'].iloc[i])
    else:
      all_list.append(all_list[i-1]+hitter_seasons[j]['홈런'].iloc[i])
  all = pd.DataFrame({
      '총홈런':all_list
  })
  hitter_seasons[j] = pd.concat([hitter_seasons[j],all], axis=1)
```

```
for j in range(5):
  all_list = []
  for i in range(len(pitcher_seasons[j])):
    if i == 0:
      all_list.append(pitcher_seasons[j]['실점'].iloc[i])
    else:
      all_list.append(all_list[i-1]+pitcher_seasons[j]['실점'].iloc[i])
  all = pd.DataFrame({
      '총실점':all_list
  })
  pitcher_seasons[j] = pd.concat([pitcher_seasons[j],all], axis=1)
pitcher_seasons[1]
```
