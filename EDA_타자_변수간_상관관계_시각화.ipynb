```
address = "./gdrive/My Drive/빅콘테스트2020/1차_최종_데이터"
union = pd.read_csv(address+"/1차_최종_공통.csv", encoding = "utf-8")
teamhitter = pd.read_csv(address+"/1차_최종_팀타자.csv", encoding = "utf-8")

union = pd.concat([union, teamhitter[['OPS','타율']]], axis=1)
```

```
team_list = []
for year in range(2016, 2021):
  a = union[(union['일자'] > '{}-05-10'.format(year)) & (union['시즌'] == year)]
  team_list.append(a)
union = pd.concat(team_list)
```

```
import plotly.express as px

fig = px.scatter(    
        x=union['평균득점'], y=union['OPS'],trendline = "ols",trendline_color_override="#ff6347"
    )


fig.update_layout(
    {
        "title": {
            "text": "Graph with go.Scatter",
            "font": {
                "size": 15
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "평균 득점"
        },
        "yaxis": {
            "title": "OPS"
        }
    }
)

fig.show()
```

```
from sklearn import linear_model

linear_regression = linear_model.LinearRegression()
linear_regression.fit(X=pd.DataFrame(union["평균득점"]), y = union["OPS"])
prediction = linear_regression.predict(X = pd.DataFrame(union["평균득점"]))
prediction
```
